```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2D tensor representing a 4x4 grayscale image
img = np.array([[0.1, 0.3, 0.4, 0.2],
                [0.6, 0.7, 0.1, 0.5],
                [0.2, 0.9, 0.8, 0.6],
                [0.3, 0.5, 0.6, 0.9]])

# Visualize the tensor as an image using Matplotlib
plt.imshow(img, cmap='gray')
plt.show()


```
